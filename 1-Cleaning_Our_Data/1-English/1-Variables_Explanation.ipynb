# Data Dictionary: XRP (Yahoo Finance) 📕

This dataset represents the "OHLCV" (Open, High, Low, Close, Volume) price history, which is the standard in financial technical analysis.

**Date:** The time indicator for the row. In Yahoo Finance, for cryptocurrencies, data is recorded daily and covers 24 hours a day (UTC).

**Ticker:** The asset's trading symbol. In your case, XRP-USD identifies the Ripple cryptocurrency against the US Dollar.

**Currency:** The monetary unit in which prices are expressed (e.g., USD).

**Open:** The price at which the first XRP transaction was made at the start of the period (00:00 UTC).

**High:** The highest price reached by the cryptocurrency during that specific day.

**Low:** The lowest price recorded during the day. It is a key indicator of support levels.

**Close:** The price of the last transaction of the day. This is the most widely used data point for calculating returns and plotting moving averages.

**Volume:** The total number of currency units (or their equivalent value in USD) exchanged during the day. High volume indicates greater liquidity and validation of the price trend.
