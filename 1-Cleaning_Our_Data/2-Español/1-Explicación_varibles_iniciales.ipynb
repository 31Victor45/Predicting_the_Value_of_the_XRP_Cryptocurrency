# Diccionario de Datos: XRP (Yahoo Finance) 📕

Este conjunto de datos representa el historial de precios "OHLCV" (Open, High, Low, Close, Volume), que es el estándar en el análisis técnico financiero.

**Date (Fecha):** Es el indicador temporal de la fila. En Yahoo Finance, para criptomonedas, los datos se registran diariamente y cubren las 24 horas del día (UTC).

**Ticker:** El símbolo de cotización del activo. En tu caso, XRP-USD identifica a la criptomoneda Ripple frente al dólar estadounidense.

**Currency (Moneda):** La unidad monetaria en la que se expresan los precios (ej. USD).

**Open (Apertura):** El precio al que se realizó la primera transacción de XRP al inicio del periodo (00:00 UTC).

**High (Máximo):** El precio más alto que alcanzó la criptomoneda durante ese día específico.

**Low (Mínimo):** El precio más bajo registrado durante la jornada. Es un gran indicador de los niveles de soporte.

**Close (Cierre):** El precio de la última transacción del día. Es el dato más utilizado para calcular rendimientos y graficar medias móviles.

**Volume (Volumen):** La cantidad total de unidades de la moneda (o su valor equivalente en USD) que se intercambiaron durante el día. Un volumen alto indica mayor liquidez y validación de la tendencia del precio.
