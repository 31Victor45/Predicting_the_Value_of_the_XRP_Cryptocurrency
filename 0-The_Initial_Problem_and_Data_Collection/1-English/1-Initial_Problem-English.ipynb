# My First Approach to the Crypto World 💹

On March 12, 2026, I was talking with some friends in my living room, and among the topics we discussed, one in particular caught my attention.

It turns out that they had invested in a cryptocurrency called XRP, and they also told me that it could be one of the currencies of the future and a good long-term investment. I have never been a big enthusiast of these things, but what does excite me is Data Science, and I particularly have a special love for time series.
